```python
import pandas as pd
import numpy as np
import torch
```

```python

npz = np.load('./train.npz')
```

```python
XTrain = npz['XTrain'].astype(np.float32)
```

```python
yTrain = npz['yTrain'].astype(np.float32)
yTrain = yTrain.reshape(len(yTrain), 1)
yTrain.shape
```

```python
device = torch.device('cuda')
class MultiLayerNet(torch.nn.Module):
  def __init__(self, D_in, H, D_out):
    super(MultiLayerNet, self).__init__()
    self.linear1 = torch.nn.Linear(D_in, H)
    self.linear2 = torch.nn.Linear(H, H)
    self.linear3 = torch.nn.Linear(H, D_out)
  def forward(self, x):
    h_relu = torch.nn.ReLU()(self.linear1(x).clamp(min=0))
    h_relu = torch.nn.ReLU()(self.linear2(h_relu).clamp(min=0))
    y_pred = self.linear3(h_relu)
    return y_pred
```

```python
XTrain.shape
```

```python
_, D_in = XTrain.shape
H = 1000
D_out = 1
```

```python
x = torch.from_numpy(XTrain[:100000]).to(device)
y = torch.from_numpy(yTrain[:100000]).to(device)
```

```python
model     = MultiLayerNet(D_in, H, D_out).to(device)
```

```python
loss_fn   = torch.nn.MSELoss(size_average=False)
```

```python
x.shape
```

```python
y.shape
```

```python

optimizer = torch.optim.Adam(model.parameters(), lr=1e-5)
for chunkTrain, chunkLabel in zip(np.array_split(XTrain, 5), np.array_split(yTrain, 5)):
    x = torch.from_numpy(chunkTrain).to(device)
    y = torch.from_numpy(chunkLabel).to(device) 
    for t in range(250):
      # Forward pass: Compute predicted y by passing x to the model
      y_pred = model(x)
      # Compute and print loss
      loss = loss_fn(y_pred, y)
      print(t, loss.item())

      # Zero gradients, perform a backward pass, and update the weights.
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()
    torch.cuda.empty_cache()
```
